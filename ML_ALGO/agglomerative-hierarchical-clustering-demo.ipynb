```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import numpy as np
```

```python
customer_data = pd.read_csv('hierarchical-clustering-with-python-and-scikit-learn-shopping-data.csv')
```

```python
customer_data.head()
```

```python
data = customer_data.iloc[:, 3:5].values
```

```python
data
```

# Get the Number of cluster using dendogram method.

```python
import scipy.cluster.hierarchy as snp

```

```python

plt.figure(figsize=(10, 7))
plt.title("Customer Dendograms")
snp.dendrogram(snp.linkage(data,method="ward"))
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
labels_=cluster.fit_predict(data)
```

```python
labels_
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(data[:,0], data[:,1], c=cluster.labels_, cmap='rainbow')
```
